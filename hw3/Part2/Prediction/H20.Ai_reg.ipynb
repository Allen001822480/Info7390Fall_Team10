```python
import h2o

h2o.init(strict_version_check=False,port=9957)
```

```python
train_df = h2o.import_file('reg_data_train.csv')
test_df = h2o.import_file('reg_data_test.csv')
```

```python
train_df.describe()
```

```python
X = list(set(list(df_train.columns)) - set(['original_interest_rate']))
Y = 'original_interest_rate'
```

```python
from h2o.automl import H2OAutoML
# Set up AutoML
aml = H2OAutoML()
```

```python
aml.train(x=X,y=Y,training_frame=train_df)
```

```python
aml_leaderboard_df=aml.leaderboard.as_data_frame()
aml_leaderboard_df
```

```python
model_set=aml_leaderboard_df['model_id']
mod_best=h2o.get_model(model_set[0])
```

```python
mod_best
```

```python
mod_perf = mod_best.model_performance(test_df)
mod_perf
```
