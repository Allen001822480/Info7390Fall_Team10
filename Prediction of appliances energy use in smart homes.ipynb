# Prediction of appliances energy use in smart homes

1. Introduction

Energy consumption in the residential sector represents an important part of the total electricity demand. In this context, a proper prediction of energy demand in housing sector is very important. The purpose of this paper is to predict the energy consumption in houses for the next 24 h, as the energy price in the day-ahead market is set for each hourly interval with one day advance.

2. The need for energy prediction in housing

Smart grid: 

Smart grid integrates the use of sensors, communications, computational ability and control in order to enhance the overall functionality of the electric power system. Smart grid initiatives seek to improve operations, maintenance and planning using modern technology in order to better manage energy use and costs.

With the smart grid demands, load management has been changing its objectives from load following to load shaping strategies, that is often control by cost. The effectiveness is shown below.

Smart home:

The smart home allows control of the services in the house through a home automation system. The electrical devices include two-way communication capabilities for allowing the energy management system to control the consumption.

Different types of predictions can be made in order to determine the energy consumption of the appliances in a house. In order to get a better load control, the energy prediction has to go down from total household energy consumption to electrical device consumption.

5. Conclusion

The paper discusses about how to evaluate the precision of a predictor in the day + 1 power management context. Different basic predictors are presented and tested for the available historical data. A relevant predictor is presented. It is based on the segmentation of data considering the patterns in energy consumption. The historical data is divided according to the results of the k-means clustering algorithm. This procedure improves the precision of the predictor.
